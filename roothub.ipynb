```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tech101 = pd.read_csv("./tech101a.csv", skipinitialspace=True)
tech101.head()
```

```python
tech101b = tech101[["MON_16_5", "Tue_17_5", "Wed_18_5", "Thur_19_5", "Fri_20_5"]]
tech101b.head()
```

```python
tech101b.T 
```

```python
tech101b = tech101b.replace(["P"], 1)
tech101b
```

```python
tech101b = tech101b.replace(["A"], 0)
tech101b
```

```python
tech101b["status"] = tech101b.sum(axis = 1)
tech101b
```

```python
tech101b["status"] = tech101b["status"].replace([5], "Certified")
tech101b["status"]
```

```python
tech101b["status"] = tech101b["status"].replace([4], "Certified")
tech101b["status"]
```

```python
tech101b["status"] = tech101b["status"].replace([3], "Certified")
tech101b["status"]
```

```python
tech101b["status"] = tech101b["status"].replace([2], "Not Certified")
tech101b["status"]
```

```python
tech101b["status"] = tech101b["status"].replace([1], "Not Certified")
tech101b["status"]
```

```python
tech101b["status"] = tech101b["status"].replace([0], "Not Certified")
tech101b["status"]
```

```python
tech101 = tech101.assign(status_confirmed = tech101b["status"])
tech101.head()
```
